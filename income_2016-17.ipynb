```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Loading the Excel file and selecting the correct sheet
file = "IncomeDistribution2010-19.xlsx"
sheet = "T3_1a"

# Read the data, skipping the first two rows (which are empty)
data = pd.read_excel(file, sheet_name=sheet, header=3)
data.head()
```

```python
# Data for 2016-17 (from the provided dataset because Student Id = 250055276)
percentiles = np.arange(1, 100)
incomes = [x for x in data['2016-17']]

print("Starting 10 incomes are -",incomes[0:10])
```

```python
# Converting income to thousands of pounds
incomes_thousands = np.array(incomes) / 1000.0

# Percentage of population with income above each value
percentage_above = 100 - percentiles
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(incomes_thousands, percentage_above, 'b-', label='Income Distribution 2016-17')
plt.xlabel('Income (£ thousands)')
plt.ylabel('Percentage of Population with Income Above (%)')
plt.title('Income Distribution: Percentage Above Income (2016-17)')
plt.grid(True)
plt.legend()
plt.xscale('linear')  # Linear scale for clarity
plt.yscale('linear')
plt.savefig('income_distribution_2016_17.png')
plt.show()
```
